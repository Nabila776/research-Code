```python
data=df.replace({
    'Yes':1,
    'No':0,
    'Positive':1,
    'Negative':0,
    'Male':1,
    'Female':0
})
data
```

```python
display(df.info(),df.head())
```

```python
plt.figure(figsize=(5,5))
plt.pie(
        data['class'].value_counts(),
        labels=['Positive','Negative'],
        colors=['SkyBlue', 'MediumPurple'],
        shadow=True,
        textprops=dict(size=15),
        explode = (0,0.05) ,
        autopct='%0.1f%%'
       )
plt.title("pie plot") 
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(['Positive','Negative'],data['class'].value_counts(),color=['lightskyblue', 'MediumPurple'])
plt.xlabel("Class") 
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(15,10))
corelation =data.corr()
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```
